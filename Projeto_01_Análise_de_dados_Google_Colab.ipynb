```python
import pandas as pd

# Load the Excel file without a header, so all rows are read as data.
df_temp = pd.read_excel(
    '08-11 - Planilha para Acompanhamento de tarefas.xlsx',
    sheet_name='Tarefas',
    header=None # Read all rows as data
)

# The actual column names are in the third row (index 2) of the Excel sheet.
# df_temp.iloc[2] contains: [NaN, 'Tarefa', 'Status', 'Hora', 'Data de início', 'Prazo', 'Observações']
header_values = df_temp.iloc[2]

# Identify valid header names (non-NaN) and their corresponding original column indices.
valid_header_indices = header_values.dropna().index
valid_header_names = header_values.dropna().tolist()

# The actual data starts from the fourth row (index 3) of the Excel sheet.
# Select all rows from index 3 onwards as the main data.
df_cleaned = df_temp.iloc[3:].copy()

# Select only the columns that correspond to the valid header names
df_cleaned = df_cleaned[valid_header_indices]

# Assign the valid header names to these columns.
df_cleaned.columns = valid_header_names

# Drop any rows that are entirely NaN (empty rows that might still be present).
df_cleaned.dropna(how='all', inplace=True)

# Display the first 10 rows of the cleaned DataFrame.
df_cleaned.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df_heat = pd.DataFrame({
    "Ana": [7, 10, 9],
    "Carlos": [6, 13, 11],
    "Beatriz": [5, 2, 3]
}, index=["Seg", "Ter", "Qua"])

plt.figure(figsize=(8, 5))
sns.heatmap(df_heat, annot=True, cmap="YlGnBu", linewidths=0.5, linecolor='gray', cbar_kws={'label': 'Tarefas Concluídas'})

plt.title("Tarefas Concluídas por Pessoa - Semana", fontsize=14, pad=12)
plt.xlabel("Responsável")
plt.ylabel("Dia da Semana")
plt.show()
```
